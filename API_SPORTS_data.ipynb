# Goalkeeper & Set Piece Analysis using API-SPORTS

This notebook demonstrates how Python can be used for sports analytics by extracting football data from the API-SPORTS platform. The focus is on goalkeeper-related performance and set piece analysis (corners and free kicks).

The workflow includes:
- API data extraction
- JSON processing
- Data cleaning and transformation
- Basic performance analysis

```python
# INSTALL & IMPORT LIBRARIES
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# API configuration
API_KEY = "387c7b2d9a864f00a8e32b547d70bfd2"
BASE_URL = "https://v3.football.api-sports.io/"
```

```python
# FETCH MATCH EVENTS
url = BASE_URL + "fixtures/events"

headers = {
    "x-apisports-key": API_KEY
}

# Fixture ID (you can change this)
params = {
    "fixture": 215662
}

response = requests.get(url, headers=headers, params=params)
data = response.json()

data
```

```python
# CONVERT JSON TO DATAFRAME
events = pd.json_normalize(data["response"])

events.head()
```

```python
# CLEAN DATA
# standardise event types
if "type" in events.columns:
    events["type"] = events["type"].astype(str).str.lower()

events.head()
```

```python
# BASIC MATCH EVENT ANALYSIS
# Count event types
event_counts = events["type"].value_counts()

event_counts
```

```python
# GOALKEEPER-RELATED ANALYSIS
# Filter goals conceded
goals = events[events["type"] == "goal"]

goals_by_team = goals.groupby("team.name").size()

goals_by_team
```

```python
# SET PIECE ANALYSIS
# Identify set piece related events (simplified proxy)
set_pieces = events[events["type"].isin(["corner", "free kick"])]

set_piece_counts = set_pieces.groupby("team.name").size()

set_piece_counts
```

```python
# VISUALISATION
if not set_piece_counts.empty:
    set_piece_counts.plot(kind="bar")
    plt.title("Set Piece Frequency by Team")
    plt.xlabel("Team")
    plt.ylabel("Count")
    plt.show()
else:
    print("No set piece events found for visualization.")
```

```python
# GOALKEEPER INSIGHT
print("Total goals in match:", len(goals))
print("\nGoals by team:")
print(goals_by_team)
```

```python
# EXPORT DATA
events.to_csv("match_events.csv", index=False)
print("File saved: match_events.csv")
```
